```python
import requests


URL_EUROPA = "https://restcountries.com/v3.1/region/europe"


resposta = requests.get(URL_EUROPA)

if resposta.status_code == 200:
    resposta_filtros = resposta.json()
    
    paises_europa = []
    for pais in resposta_filtros:
        novo_pais = {
            "nome": pais["name"]["common"],
            "capital": pais["capital"][0] if "capital" in pais and pais["capital"] else None,
            "populacao": pais["population"]
        }
        paises_europa.append(novo_pais)
    
    
    print(paises_europa)
else:
    print(f"Erro na requisi√ß√£o: {resposta.status_code}")
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt


URL_EUROPA = "https://restcountries.com/v3.1/region/europe"


resposta = requests.get(URL_EUROPA)

if resposta.status_code == 200:
    resposta_filtros = resposta.json()
    
    paises_europa = []
    for pais in resposta_filtros:
        novo_pais = {
            "nome": pais["name"]["common"],
            "capital": pais["capital"][0] if "capital" in pais and pais["capital"] else None,
            "populacao": pais["population"]
        }
        paises_europa.append(novo_pais)
    
   
    df_europa = pd.DataFrame(paises_europa)

   
    df_europa = df_europa.sort_values(by='populacao', ascending=False)

    
    plt.figure(figsize=(15, 10))
    plt.barh(df_europa['nome'], df_europa['populacao'], color='skyblue')
    plt.xlabel('Popula√ß√£o')
    plt.ylabel('Pa√≠s')
    plt.title('Popula√ß√£o dos pa√≠ses da Europa')
    plt.gca().invert_yaxis()  # Inverte o eixo Y para mostrar o mais populoso no topo
    plt.show()
else:
    print(f"Erro na requisi√ß√£o: {resposta.status_code}")
```
